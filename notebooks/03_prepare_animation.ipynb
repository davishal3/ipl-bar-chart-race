```python
# Computing cumulative wins
import pandas as pd
import numpy as np

CLEAN_CSV = "../data/processed/ipl_matches_clean.csv"

df = pd.read_csv(CLEAN_CSV)
df.head()
```

```python
# Wins per season

wins = (
    df.groupby(["season", "winner"])
    .size()
    .reset_index(name="wins")
    .rename(columns={"winner": "team"})
)
```

```python
# create full grid

seasons = sorted(df["season"].unique())
teams = sorted(df["winner"].unique())

grid = pd.MultiIndex.from_product([seasons, teams], names=["season", "team"]).to_frame(index=False)

grid = grid.merge(wins, on=["season", "team"], how="left").fillna(0)
wins["wins"] = wins["wins"].astype(int)
```

```python
# cumulative wins

wins = wins.sort_values(["team", "season"])
wins["cum_wins"] = wins.groupby("team")["wins"].cumsum()
wins.shape
wins.info()
wins.head()
```

```python
OUT_CSV = "../data/processed/ipl_cumulative_wins.csv"
wins.to_csv(OUT_CSV, index=False)
print("Saved:", OUT_CSV)
```

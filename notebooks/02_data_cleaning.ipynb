```python
# Fixing Team name changes

import pandas as pd

RAW_CSV = "../data/processed/ipl_matches_raw.csv"
CLEAN_CSV = "../data/processed/ipl_matches_clean.csv"

df = pd.read_csv(RAW_CSV)
df.head()
```

```python
team_map = {
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Rising Pune Supergiant": "Rising Pune Supergiants"
}

df["winner"] = df["winner"].replace(team_map)
```

```python
# Saving clean csv
df.to_csv(CLEAN_CSV, index=False)
print("Saved:", CLEAN_CSV)
```
